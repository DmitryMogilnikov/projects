#### 4. Выведите таблицу, которая позволит просмотреть первую дату поставки и общее количество поставленной вакцины в рамках Марки вакцины и страны.

```python
parse_data = df.groupby(['Страна', 'Марка вакцины']).agg(first_date = pd.NamedAgg(column="Дата поставки/новости", aggfunc="min"), 
                                                    sum_vaccine = pd.NamedAgg(column="Вакцина (кол-во)", aggfunc="sum"))
```

```python
print(parse_data)
```

#### 6. Достать домены источников 1, 2, 3
Для столбцов Источников создать столбцы Доменов, где будет извлечен только домен (источник: https://sputnik-abkhazia.ru/Abkhazia/20200320/..., домен: sputnik-abkhazia.ru). В исходном датасете 3 столбца источника и соответственно должно быть создана 3 столбца с доменами.

```python
df.head()
```

```python
import numpy as np
from urllib.parse import urlparse

for num in range(1, 4):
    df[f"Источник {num}"] = df[f"Источник {num}"].fillna("https:/test.com/123")
    pos = df.columns.get_loc(f"Источник {num}") + 1
    domain = df[f"Источник {num}"].apply(lambda domain: urlparse(domain).netloc)
    df.insert(pos, f"Домен {num}", domain)
```

```python
df = df.replace("https:/test.com/123", "")
```
